```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
data=pd.read_csv(r"C:\Users\VIRAJ\Downloads\Fish.csv")
data
```

```python
data.Species.value_counts()
```

```python
data.columns
```

```python
fv=data.drop(["Weight"],axis=1)
fv
```

```python
cl=data[["Weight"]]
cl
```

```python
x_train,x_test,y_train,y_test=train_test_split(fv,cl,train_size=0.7)
x_train
```

```python
ohe=OneHotEncoder()
d=ohe.fit_transform(x_train[["Species"]])
df=pd.DataFrame(d.toarray(),columns=ohe.categories_)
df
```

```python
x_train=x_train.drop(["Species"],axis=1)
x_train.reset_index(drop=True,inplace=True)
x_train
```

```python
new_x_train=pd.concat([df,x_train],axis=1)
new_x_train
```

```python
d=ohe.transform(x_test[["Species"]])
d.toarray()
```

```python
df1=pd.DataFrame(d.toarray(),columns=ohe.categories_)
df1
```

```python
x_test=x_test.drop(["Species"],axis=1)
x_test.reset_index(drop=True,inplace=True)
x_test
```

```python
new_x_test=pd.concat([df1,x_test],axis=1)
new_x_test
```

```python
lr=LinearRegression()
model=lr.fit(new_x_train,y_train)
predict=model.predict(new_x_test)
predict
```

```python
mean_squared_error(y_test,predict)
```

```python
mean_absolute_error(y_test,predict)
```

```python
r2_score(y_test,predict)
```
